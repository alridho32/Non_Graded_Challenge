```python
class Book:
    def __init__(self, title: str, author: str, id):
        self.title = title
        self.author = author
        self.unique_id = id

class Library: 
    def __init__(self):
        self.catalog = []

    def addBook(self, title, author, id):
        book = Book(title, author, id)
        self.catalog.append(book)
        print(f'buku {title} telah ditambahkan')

    def searchBook(self, title= None, author = None):
        if title is not None:
            for i in self.catalog:
                if title.lower() == i.title.lower():
                    result = []
                    result.append({'title': i.title,
                                   'author': i.author,
                                   'id': i.unique_id})
            return result
            for i in result:
                print(i)
                print('---'*10)

        elif author is not None:
            for i in self.catalog:
                if author == i.author:
                    print('title:', i.title)
                    print('author: ', i.author)
                    print('id: ', i.unique_id)
                    print('---'*10)

    def removeBook(self, title):
        for i in self.catalog:
            if title == i.title:
                self.catalog.remove(i)

    def displayBooks(self):
        for i in self.catalog:
            print('title:', i.title)
            print('author: ', i.author)
            print('id: ', i.unique_id)
            print('---'*10)

```
